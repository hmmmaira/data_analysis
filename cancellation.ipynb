```python
data = data[["no_of_adults","no_of_children","no_of_weekend_nights","no_of_week_nights","type_of_meal_plan","room_type_reserved","booking_status","avg_price_per_room","no_of_previous_cancellations"]]
# making a new dataset with only desired columns from previous dataset
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
# allows to transform data to useable data , i.e. change it to numeric values/ dummies to allow for analysis
```

```python
data["type_of_meal_plan"].head()
```

```python
data["type_of_meal_plan"] = labelencoder.fit_transform(data["type_of_meal_plan"])
# allows to transform data to useable data , i.e. change it to numeric values/ dummies to allow for analysis
```

```python
data["type_of_meal_plan"].head()    # checking column data head
```

```python
data["booking_status"] = labelencoder.fit_transform(data["booking_status"])          
data["room_type_reserved"] = labelencoder.fit_transform(data["room_type_reserved"])
# allows to transform data to useable data , i.e. change it to numeric values/ dummies to allow for analysis
```

```python
x = data[["no_of_adults","no_of_children","no_of_weekend_nights","no_of_week_nights","type_of_meal_plan","room_type_reserved","no_of_previous_cancellations","avg_price_per_room"]]
y = data[["booking_status"]]
# segregating the dependant and independant variables
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
# divide the data set into train and test 
# use the train data to train the model and test data to check how well the model fits
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.fit_transform(x_test)
# scaler is used to scale the data to better handle the data 
```

```python
from sklearn.metrics import accuracy_score
def model_results(predictions):
    print(f"Accuracy score of the model is  {accuracy_score(y_test,predictions)}")

# accuracy score to see how well the model fits 
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_x_train,y_train)
```

```python
log_predictions = log_model.predict(scaled_x_test)
model_results(log_predictions)
```
