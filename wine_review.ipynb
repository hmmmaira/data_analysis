```python
def stars(rows):
    if rows.country == 'Canada':
        return 3
    elif rows.points >= 95:
        return 3
    elif rows.points >= 85:
        return 2
    else:
        return 1
star_ratings = data.apply(stars,axis = 'columns')
star_ratings
```
